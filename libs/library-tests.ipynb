```python
# define bounding box
lat_bottom, lat_top = 33.6, 34.3
lon_bottom, lon_top = -118.6, -117.9
extent = (lon_bottom, lon_top, lat_bottom, lat_top)

# input data shape
dim = 40
frames_per_sample = 5

# date range of data
start_date, end_date = "2025-01-10-00", "2025-01-17-00"
```

```python
from airnowdata import AirNowData

AD = AirNowData(
    start_date=start_date,
    end_date=end_date,
    extent=extent,
    airnow_api_key=None,
    save_dir="../data/airnow.json",
    frames_per_sample=frames_per_sample,
    dim=dim
)
```

```python
X_airnow = AD.data
print(X_airnow.shape)
print(AD.air_sens_loc)
print(AD.ground_site_grids[0])
for x, y in AD.air_sens_loc:
    print(AD.ground_site_grids[0][x][y], AD.air_sens_loc[(x, y)])
```

```python
HRRRData = HRRRData(
    start_date=start_date,
    end_date=end_date,
    extent=extent,
    product='MASSDEN',
    frames_per_sample=frames_per_sample,
    dim=dim,
    verbose=True
)
```
