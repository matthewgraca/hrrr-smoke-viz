```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sys
import numpy as np

BASE_DIR = '/home/mgraca/Workspace/hrrr-smoke-viz'
RESULTS_DIR = f'{BASE_DIR}/pwwb-experiments/tensorflow/multipath_archive/results/Two-Path-test'
CACHE_DIR = f'{BASE_DIR}/pwwb-experiments/tensorflow/final_input_data/multipath_data/25_scale/npy_files'
sys.path.append(BASE_DIR)
```

```python
y_pred = np.load(f'{RESULTS_DIR}/Y_pred.npy')
y_test = np.load(f'{CACHE_DIR}/Y_grid_test.npy')
y_pred.shape, y_test.shape
```

```python
idx = 1040
def sample_plotter(images, idx, rows=4, cols=5):
    fig, axes = plt.subplots(rows, cols, figsize=(14, 10))
    for r in range(axes.shape[0]):
        for c in range(axes.shape[1]):
            im = axes[r][c].imshow(images[idx, c])
            axes[r][c].axis('off') 
            axes[r][c].set_title(f"Sample {idx}, frame {c}", fontsize=8)
            fig.colorbar(im, ax=axes[r][c], shrink=0.8)
        idx += 1

    plt.tight_layout()
    plt.show()
sample_plotter(y_pred, idx)
```

```python
sample_plotter(y_test, idx)
```

```python
x_test = np.load(f'{RESULTS_DIR}/AR_0_X_test.npy')
sample_plotter(x_test[..., 0], idx)
```

```python
X_test_files = [f'{RESULTS_DIR}/AR_{i}_X_test.npy' for i in range(5)]
def sample_channel_plotter(files, channel, idx):
    fig, axes = plt.subplots(5, 5, figsize=(14, 10))
    for r, X_test in zip(range(axes.shape[0]), files):
        images = np.load(X_test)[..., channel]
        for c in range(axes.shape[1]):
            im = axes[r][c].imshow(images[idx, c])
            axes[r][c].axis('off') 
            axes[r][c].set_title(f"Sample {idx}, frame {c}", fontsize=8)
            fig.colorbar(im, ax=axes[r][c], shrink=0.8)
    
    plt.tight_layout()
    plt.show()

sample_channel_plotter(X_test_files, channel=0, idx=idx)
```
