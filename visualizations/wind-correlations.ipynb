# Baseline - past vs future pm2.5

```python
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.append('/home/mgraca/Workspace/hrrr-smoke-viz')
filedir = '/home/mgraca/Workspace/hrrr-smoke-viz/pwwb-experiments/tensorflow/final_input_data/two_years_sat_data'

x_raw = np.load(f'{filedir}/AirNow_PM25_X_train.npy')
y_raw = np.load(f'{filedir}/Y_train.npy')

x = np.mean(x_raw, axis=(1, 2, 3))
y = np.mean(y_raw, axis=(1, 2))

x.shape, y.shape
```

```python
plt.scatter(x, y)
plt.title("Relationship between Past PM2.5 and Future PM 2.5")
plt.xlabel("Past PM2.5")
plt.ylabel("Future PM2.5")

m, b = np.polyfit(x, y, 1)
plt.plot(x, m*x + b, color='red', linestyle=':', alpha=0.75)

plt.show()
```
